# Prepare rules for the all the data sets 
1) Try different values of support and confidence. Observe the change in number of rules for different support,confidence values
2) Change the minimum length in apriori algorithm
3) Visulize the obtained rules using different plots 

```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori,association_rules
import warnings
import matplotlib.pyplot as plt
warnings.filterwarnings("ignore")
```

```python
data=pd.read_csv("Association Rules_Book.csv.xls")
data.head(5)
```

```python
fq_itemsets=apriori(data,min_support=0.1,use_colnames=True)
fq_itemsets
```

```python
fq_itemsets1=apriori(data,min_support=0.15,use_colnames=True)
fq_itemsets1
```

```python
fq_itemsets2=apriori(data,min_support=0.25,use_colnames=True)
fq_itemsets2
```

```python
rules=association_rules(fq_itemsets, metric='lift',min_threshold=1.2)
rules
```

```python
rules=association_rules(fq_itemsets, metric='lift',min_threshold=2)
rules
```

```python
max(rules['antecedent support'])
```

```python
rules1=association_rules(fq_itemsets, metric='lift',min_threshold=1.3)
rules1
```

```python
max(rules1['antecedent support'])
```

```python
rules2=association_rules(fq_itemsets1, metric='lift',min_threshold=1.3)
rules2
```

```python
rules3=association_rules(fq_itemsets2, metric='lift',min_threshold=1.3)
rules3
```

```python
plt.scatter(rules2['support'],rules2['confidence'])
plt.xlabel("Confidence")
plt.ylabel("Support")
plt.show
```
