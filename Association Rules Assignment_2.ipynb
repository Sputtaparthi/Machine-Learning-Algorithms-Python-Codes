# Prepare rules for the all the data sets 
1) Try different values of support and confidence. Observe the change in number of rules for different support,confidence values
2) Change the minimum length in apriori algorithm
3) Visulize the obtained rules using different plots 

```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori,association_rules
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("my_movies.csv.xls")
print(df.head(10))
df.shape
```

```python
data=df.iloc[:,5:]
data
```

```python
for i in data.columns:
    print(i)
    print(data[i].value_counts())
    print()
    
```

```python
fq_itemsets=apriori(data,min_support=0.1,use_colnames=True)
fq_itemsets1=apriori(data,min_support=0.15,use_colnames=True)
fq_itemsets2=apriori(data,min_support=0.2,use_colnames=True)
fq_itemsets3=apriori(data,min_support=0.25,use_colnames=True)
fq_itemsets
```

```python
fq_itemsets1
```

```python
fq_itemsets2
```

```python
fq_itemsets3
```

```python
rule=association_rules(fq_itemsets1,metric='lift',min_threshold=1.2)
rule1=association_rules(fq_itemsets1,metric='lift',min_threshold=1.5)
rule2=association_rules(fq_itemsets1,metric='lift',min_threshold=1.7)
rule3=association_rules(fq_itemsets1,metric='lift',min_threshold=2)
```

```python
plt.scatter(rule['support'],rule['confidence'])
plt.xlabel("Confidence")
plt.ylabel("Support")
plt.show
```
